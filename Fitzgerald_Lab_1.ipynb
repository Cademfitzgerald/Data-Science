```python
plt.plot(x,y,marker ='o', linestyle ='None', markerfacecolor ='#D89F62',label = 'Graph of x squared')
plt.title('My First Graph');
plt.xlabel("X axis label goes here");
plt.ylabel('Y axis label goes here');
plt.legend();
```

```python
x=np.linspace(0,2,100)
plt.plot(x,x, "g--", label="linear")
plt.plot(x,x**2, "ms", label="squared")
plt.plot(x,x**3, "b^", label="cubed")
plt.legend();
plt.xlabel("X axis label goes here");
plt.ylabel('Y axis label goes here');
```
